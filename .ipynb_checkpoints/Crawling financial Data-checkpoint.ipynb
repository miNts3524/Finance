```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import pandas as pd
import mplfinance as mpf
import matplotlib.pyplot as plt
```

```python
url =('https://finance.naver.com/item/sise_day.nhn?code=005930&page=1')
url_kakao=('https://finance.naver.com/item/sise_day.nhn?code=035720&page=1')
user_agent=('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36')
```

Get the final Page

```python
with requests.get(url,headers={'User-agent':user_agent}) as doc:
    html = BeautifulSoup(doc.text,'lxml') # method of pasing
    pgrr=html.find('td',class_= 'pgRR')  #page Right Right
    s=str(pgrr.a['href']).split('=')  #return href of a tag
    last_page = s[-1]
    print(last_page)
```

```python
with requests.get(url_kakao,headers={'User-agent':user_agent}) as doc:
    html = BeautifulSoup(doc.text,'lxml')
    pgrr=html.find('td',class_='pgRR')
    l=str(pgrr.a['href']).split('=')
    last_page_k=l[-1]
    print(last_page_k)
```

```python
df=pd.DataFrame()

url_2 = 'https://finance.naver.com/item/sise_day.nhn?code=005930'
for page in range(1,int(last_page)+1):
    page_url = '{}&page={}'.format(url_2,page)
    response_page=requests.get(page_url,headers={'User-agent':user_agent}).text
    df=df.append(pd.read_html(response_page)[0])
```

```python
df2=pd.DataFrame()

url_k2='https://finance.naver.com/item/sise_day.nhn?code=035720'
for page in range(int(last_page_k)+1):
    page_k_url = '{}&page={}'.format(url_k2,page)
    response_page_k=requests.get(page_k_url,headers={'User-agent':user_agent}).text
    df= df.append(pd.read_html(response_page)[0])
    
```

```python
df = df.dropna() # remove NaN
df=df.reset_index(drop=True)  # reset index

df2 = df2.dropna()
df2=df2.reset_index(drop=True)
```

```python

new_df = df[0:200]
new_df = new_df.sort_values(by='날짜')
```
