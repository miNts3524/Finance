```python
import pandas as pd  # pd has series & dataframe
```

```python
import matplotlib.pyplot as plt
```

```python
from pandas_datareader import data as pdr
import yfinance as yf
```

```python
yf.pdr_override()

sec=pdr.get_data_yahoo('005930.KS',start='2018-05-04')
msft=pdr.get_data_yahoo('MSFT',start='2018-05-04')
g_stop=pdr.get_data_yahoo('GME',start='2018-05-04')
Tesla=pdr.get_data_yahoo('TSLA',start='2018-05-04')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(sec.index,sec.Close,'b',label='Samsung Elctronics')
plt.plot(msft.index,msft.Close,'r',label='Microsoft')
plt.plot(g_stop.index,g_stop.Close,label='GameStop')
plt.plot(Tesla.index,Tesla.Close,label='Tesla')
plt.legend(loc='best')
plt.show()
```

```python
sec_dpc = (sec['Close']/sec['Close'].shift(1)-1)*100
sec_dpc[0]=0
sec_dpc= sec_dpc.cumsum()

msft_dpc = (msft['Close']/msft['Close'].shift(1)-1)*100
msft_dpc[0]=0
msft_dpc=msft_dpc.cumsum()

g_stop_dpc = (g_stop['Close']/g_stop['Close'].shift(1)-1)*100
g_stop_dpc[0]=0
g_stop_dpc=g_stop_dpc.cumsum()

Tesla_dpc=(Tesla['Close']/Tesla['Close'].shift(1)-1)*100
Tesla_dpc[0]=0
Tesla_dpc=Tesla_dpc.cumsum()
```

```python
plt.plot(sec.index,sec_dpc,'b',label="Samsung Electronics")
plt.plot(msft.index,msft_dpc,'r',label="Microsoft")
plt.plot(g_stop.index,g_stop_dpc,'g',label="GameStop")
plt.plot(Tesla.index,Tesla_dpc,'y',label='Tesla')

plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
plt.hist(Tesla_dpc,bins=30)
plt.grid(True)
plt.show()
```
